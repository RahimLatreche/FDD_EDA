```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
from scipy import stats
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')

# Set style for better-looking plots
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# ========================================
# PHASE 1: PRE-CYCLING ANALYSIS (PRIORITY)
# Target Events: June 4-26, 2024 (before official detection)
# ========================================

print("="*80)
print("PHASE 1: PRE-CYCLING SENSOR DATA ANALYSIS")
print("Target Period: June 4-26, 2024 (Before Official Detection)")
print("="*80)

# Load the merged sensor data
print("Loading sensor data...")
df_sensors = pd.read_csv('comerica_rtu_merged_data.csv')

# Convert timestamp to datetime
df_sensors['timestamp'] = pd.to_datetime(df_sensors['timestamp'])

# Define analysis periods
pre_cycling_start = datetime(2024, 6, 1)
pre_cycling_end = datetime(2024, 6, 30)
first_cycling_event = datetime(2024, 6, 4, 14, 0)  # June 4, 14:00

# Known cycling event dates from our analysis
cycling_events = [
    datetime(2024, 6, 4, 14, 0),   # First event
    datetime(2024, 6, 17, 12, 56), # Second event
    datetime(2024, 6, 21, 10, 50), # Third event
    datetime(2024, 6, 26, 12, 1)   # Fourth event
]

print(f"Analysis Period: {pre_cycling_start.date()} to {pre_cycling_end.date()}")
print(f"Cycling Events in Period: {len(cycling_events)}")

# Filter data for pre-cycling analysis period
df_period = df_sensors[
    (df_sensors['timestamp'] >= pre_cycling_start) & 
    (df_sensors['timestamp'] <= pre_cycling_end)
].copy()

print(f"Sensor records in analysis period: {len(df_period):,}")

# ========================================
# IDENTIFY COOLING-RELATED SENSORS
# ========================================
print("\n" + "="*60)
print("IDENTIFYING COOLING-RELATED SENSORS")
print("="*60)

# Find all unique sensor types (navName)
unique_sensors = df_period['navName'].unique()
print(f"Total unique sensors: {len(unique_sensors)}")

# Filter for cooling-related sensors
cooling_keywords = [
    'cooling', 'cool', 'stage', 'demand', 'setpoint', 'temperature', 
    'temp', 'discharge', 'supply', 'return', 'zone', 'command', 'enable'
]

cooling_sensors = []
for sensor in unique_sensors:
    if any(keyword.lower() in sensor.lower() for keyword in cooling_keywords):
        cooling_sensors.append(sensor)

print(f"Cooling-related sensors found: {len(cooling_sensors)}")
print("\nCooling sensor types:")
for sensor in sorted(cooling_sensors):
    count = len(df_period[df_period['navName'] == sensor])
    print(f"  • {sensor}: {count:,} readings")

# ========================================
# VISUALIZATION 1: SENSOR DATA AVAILABILITY
# ========================================
print("\n" + "="*60)
print("CREATING VISUALIZATIONS")
print("="*60)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Phase 1: Pre-Cycling Analysis - Sensor Overview', fontsize=16, fontweight='bold')

# Plot 1: Sensor data availability
sensor_counts = []
sensor_names = []
for sensor in sorted(cooling_sensors):
    count = len(df_period[df_period['navName'] == sensor])
    sensor_counts.append(count)
    sensor_names.append(sensor.replace(' ', '\n'))  # Break long names

ax1.barh(range(len(sensor_names)), sensor_counts, color='skyblue')
ax1.set_yticks(range(len(sensor_names)))
ax1.set_yticklabels(sensor_names, fontsize=8)
ax1.set_xlabel('Number of Readings')
ax1.set_title('Sensor Data Availability (June 2024)')
ax1.grid(axis='x', alpha=0.3)

# Plot 2: Cycling events timeline
cycling_dates = [event.date() for event in cycling_events]
cycling_times = [event.hour + event.minute/60 for event in cycling_events]

ax2.scatter(cycling_dates, cycling_times, s=200, c='red', alpha=0.7, edgecolors='darkred')
for i, event in enumerate(cycling_events):
    ax2.annotate(f'Event {i+1}', 
                (cycling_dates[i], cycling_times[i]), 
                xytext=(5, 5), textcoords='offset points', fontsize=8)
ax2.set_ylabel('Time of Day (24hr)')
ax2.set_title('Cycling Events Timeline (June 2024)')
ax2.grid(True, alpha=0.3)
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45)

# Plot 3: Temperature overview
temp_sensors = [s for s in cooling_sensors if 'temperature' in s.lower() or 'temp' in s.lower()]
for temp_sensor in temp_sensors:
    temp_data = df_period[df_period['navName'] == temp_sensor].copy()
    temp_data['value_numeric'] = pd.to_numeric(temp_data['value'], errors='coerce')
    temp_data = temp_data.dropna().sort_values('timestamp')
    
    if len(temp_data) > 0:
        ax3.plot(temp_data['timestamp'], temp_data['value_numeric'], 
                alpha=0.7, label=temp_sensor, linewidth=1)

# Mark cycling events on temperature plot
for event in cycling_events:
    ax3.axvline(x=event, color='red', linestyle='--', alpha=0.7)

ax3.set_ylabel('Temperature (°F)')
ax3.set_title('Temperature Trends vs Cycling Events')
ax3.legend(fontsize=8)
ax3.grid(True, alpha=0.3)
plt.setp(ax3.xaxis.get_majorticklabels(), rotation=45)

# Plot 4: Cooling demand overview
demand_data = df_period[df_period['navName'] == 'Cooling Demand'].copy()
demand_data['value_numeric'] = pd.to_numeric(demand_data['value'], errors='coerce')
demand_data = demand_data.dropna().sort_values('timestamp')

if len(demand_data) > 0:
    ax4.plot(demand_data['timestamp'], demand_data['value_numeric'], 
             color='blue', alpha=0.7, linewidth=1)
    
    # Mark cycling events
    for event in cycling_events:
        ax4.axvline(x=event, color='red', linestyle='--', alpha=0.7)
    
    # Add trend line
    x_numeric = np.arange(len(demand_data))
    z = np.polyfit(x_numeric, demand_data['value_numeric'], 1)
    p = np.poly1d(z)
    ax4.plot(demand_data['timestamp'], p(x_numeric), "r--", alpha=0.8, linewidth=2)

ax4.set_ylabel('Cooling Demand (%)')
ax4.set_title('Cooling Demand Trend vs Cycling Events')
ax4.grid(True, alpha=0.3)
plt.setp(ax4.xaxis.get_majorticklabels(), rotation=45)

plt.tight_layout()
plt.savefig('01_sensor_overview_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# SENSOR FOCUS ANALYSIS
# ========================================

def analyze_sensor_pattern(df, sensor_name, period_label=""):
    """Analyze patterns for a specific sensor"""
    sensor_data = df[df['navName'] == sensor_name].copy()
    if len(sensor_data) == 0:
        return None
    
    # Convert value to numeric, handling any string values
    sensor_data['value_numeric'] = pd.to_numeric(sensor_data['value'], errors='coerce')
    sensor_data = sensor_data.dropna(subset=['value_numeric'])
    
    if len(sensor_data) == 0:
        return None
    
    # Sort by timestamp
    sensor_data = sensor_data.sort_values('timestamp')
    
    # Calculate basic statistics
    stats_dict = {
        'sensor_name': sensor_name,
        'period': period_label,
        'count': len(sensor_data),
        'mean': sensor_data['value_numeric'].mean(),
        'std': sensor_data['value_numeric'].std(),
        'min': sensor_data['value_numeric'].min(),
        'max': sensor_data['value_numeric'].max(),
        'range': sensor_data['value_numeric'].max() - sensor_data['value_numeric'].min()
    }
    
    # Calculate variability metrics
    if len(sensor_data) > 1:
        # Rate of change
        sensor_data['time_diff'] = sensor_data['timestamp'].diff().dt.total_seconds() / 60  # minutes
        sensor_data['value_diff'] = sensor_data['value_numeric'].diff()
        sensor_data['rate_of_change'] = sensor_data['value_diff'] / sensor_data['time_diff']
        
        # Volatility metrics
        stats_dict['volatility'] = sensor_data['rate_of_change'].std()
        stats_dict['max_change_rate'] = sensor_data['rate_of_change'].abs().max()
        
        # Cycling detection (for stage commands)
        if 'command' in sensor_name.lower() or 'stage' in sensor_name.lower():
            # Count state changes
            state_changes = (sensor_data['value_numeric'].diff() != 0).sum()
            stats_dict['state_changes'] = state_changes
            stats_dict['cycling_frequency'] = state_changes / (len(sensor_data) / 60)  # changes per hour
    
    return sensor_data, stats_dict

print("\n" + "="*60)
print("SENSOR FOCUS ANALYSIS")
print("="*60)

# Categorize sensors
demand_sensors = [s for s in cooling_sensors if 'demand' in s.lower()]
stage_sensors = [s for s in cooling_sensors if 'stage' in s.lower() and 'command' in s.lower()]
setpoint_sensors = [s for s in cooling_sensors if 'setpoint' in s.lower()]
temp_sensors = [s for s in cooling_sensors if 'temperature' in s.lower() or 'temp' in s.lower()]

# Store analysis results
demand_analysis = {}
stage_analysis = {}
setpoint_analysis = {}
temp_analysis = {}

# 1. COOLING DEMAND ANALYSIS
print("\n### 1. COOLING DEMAND ANALYSIS")
print("-" * 40)
print(f"Demand sensors found: {demand_sensors}")

for sensor in demand_sensors:
    result = analyze_sensor_pattern(df_period, sensor, "Pre-Cycling Period")
    if result:
        sensor_data, stats = result
        demand_analysis[sensor] = {'data': sensor_data, 'stats': stats}
        
        print(f"\n{sensor}:")
        print(f"  • Mean: {stats['mean']:.2f}, Std: {stats['std']:.2f}")
        print(f"  • Range: {stats['min']:.2f} to {stats['max']:.2f}")
        if 'volatility' in stats:
            print(f"  • Volatility: {stats['volatility']:.4f}")

# 2. COOLING STAGE COMMANDS ANALYSIS
print("\n### 2. COOLING STAGE COMMANDS ANALYSIS")
print("-" * 40)
print(f"Stage command sensors found: {stage_sensors}")

for sensor in stage_sensors:
    result = analyze_sensor_pattern(df_period, sensor, "Pre-Cycling Period")
    if result:
        sensor_data, stats = result
        stage_analysis[sensor] = {'data': sensor_data, 'stats': stats}
        
        print(f"\n{sensor}:")
        print(f"  • Unique values: {sorted(sensor_data['value_numeric'].unique())}")
        print(f"  • Mean runtime: {stats['mean']:.2f}")
        if 'state_changes' in stats:
            print(f"  • State changes: {stats['state_changes']}")
            print(f"  • Cycling frequency: {stats['cycling_frequency']:.2f} changes/hour")

# 3. SETPOINT SENSORS ANALYSIS  
print("\n### 3. SETPOINT SENSORS ANALYSIS")
print("-" * 40)
print(f"Setpoint sensors found: {setpoint_sensors}")

for sensor in setpoint_sensors:
    result = analyze_sensor_pattern(df_period, sensor, "Pre-Cycling Period")
    if result:
        sensor_data, stats = result
        setpoint_analysis[sensor] = {'data': sensor_data, 'stats': stats}
        
        print(f"\n{sensor}:")
        print(f"  • Mean setpoint: {stats['mean']:.2f}")
        print(f"  • Setpoint range: {stats['min']:.2f} to {stats['max']:.2f}")
        print(f"  • Stability (std): {stats['std']:.2f}")
        if 'volatility' in stats:
            print(f"  • Change rate volatility: {stats['volatility']:.4f}")

# 4. TEMPERATURE SENSORS ANALYSIS
print("\n### 4. TEMPERATURE SENSORS ANALYSIS")
print("-" * 40)
print(f"Temperature sensors found: {temp_sensors}")

for sensor in temp_sensors:
    result = analyze_sensor_pattern(df_period, sensor, "Pre-Cycling Period")
    if result:
        sensor_data, stats = result
        temp_analysis[sensor] = {'data': sensor_data, 'stats': stats}
        
        print(f"\n{sensor}:")
        print(f"  • Mean temperature: {stats['mean']:.2f}°F")
        print(f"  • Temperature range: {stats['min']:.2f}°F to {stats['max']:.2f}°F")
        print(f"  • Temperature stability: {stats['std']:.2f}°F")
        if 'volatility' in stats:
            print(f"  • Temperature volatility: {stats['volatility']:.4f}°F/min")

# ========================================
# VISUALIZATION 2: DETAILED SENSOR ANALYSIS
# ========================================

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Phase 1: Detailed Sensor Analysis - June 2024', fontsize=16, fontweight='bold')

# Plot 1: Cooling Demand Detailed View
ax1 = axes[0, 0]
if 'Cooling Demand' in demand_analysis:
    cooling_demand_data = demand_analysis['Cooling Demand']['data']
    ax1.plot(cooling_demand_data['timestamp'], cooling_demand_data['value_numeric'], 
             color='blue', alpha=0.7, linewidth=1)
    
    # Mark cycling events
    for i, event in enumerate(cycling_events):
        ax1.axvline(x=event, color='red', linestyle='--', alpha=0.7)
        ax1.text(event, ax1.get_ylim()[1]*0.9, f'C{i+1}', 
                rotation=90, fontsize=8, ha='center')
    
    # Add mean line
    mean_demand = cooling_demand_data['value_numeric'].mean()
    ax1.axhline(y=mean_demand, color='green', linestyle='-', alpha=0.7, 
                label=f'Mean: {mean_demand:.1f}%')
    
    ax1.set_ylabel('Cooling Demand (%)')
    ax1.set_title('Cooling Demand vs Cycling Events')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

# Plot 2: Setpoint Analysis
ax2 = axes[0, 1]
for sensor_name, data in setpoint_analysis.items():
    if 'Occupied' in sensor_name:  # Focus on occupied setpoints
        sensor_data = data['data']
        ax2.plot(sensor_data['timestamp'], sensor_data['value_numeric'], 
                alpha=0.7, label=sensor_name.replace('Occupied ', ''), linewidth=1)

# Mark cycling events
for event in cycling_events:
    ax2.axvline(x=event, color='red', linestyle='--', alpha=0.5)

ax2.set_ylabel('Setpoint (°F)')
ax2.set_title('Occupied Setpoints vs Cycling Events')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Temperature Correlation
ax3 = axes[1, 0]
# Plot Zone vs Outside Temperature
if 'Zone Temperature' in temp_analysis and 'Outside Temperature' in temp_analysis:
    zone_data = temp_analysis['Zone Temperature']['data']
    outside_data = temp_analysis['Outside Temperature']['data']
    
    ax3.plot(outside_data['timestamp'], outside_data['value_numeric'], 
             alpha=0.7, label='Outside Temp', color='orange', linewidth=1)
    ax3_twin = ax3.twinx()
    ax3_twin.plot(zone_data['timestamp'], zone_data['value_numeric'], 
                  alpha=0.7, label='Zone Temp', color='blue', linewidth=1)
    
    # Mark cycling events
    for event in cycling_events:
        ax3.axvline(x=event, color='red', linestyle='--', alpha=0.5)
    
    ax3.set_ylabel('Outside Temp (°F)', color='orange')
    ax3_twin.set_ylabel('Zone Temp (°F)', color='blue')
    ax3.set_title('Temperature Correlation vs Cycling Events')
    ax3.grid(True, alpha=0.3)

# Plot 4: Volatility Heatmap
ax4 = axes[1, 1]
volatility_data = []
sensor_labels = []

for category, analysis_dict in [
    ('Demand', demand_analysis),
    ('Setpoints', setpoint_analysis),
    ('Temperature', temp_analysis)
]:
    for sensor, data in analysis_dict.items():
        if 'volatility' in data['stats']:
            volatility_data.append(data['stats']['volatility'])
            sensor_labels.append(sensor.replace(' ', '\n'))

if volatility_data:
    # Create horizontal bar chart for volatility
    colors = plt.cm.viridis(np.linspace(0, 1, len(volatility_data)))
    bars = ax4.barh(range(len(sensor_labels)), volatility_data, color=colors)
    ax4.set_yticks(range(len(sensor_labels)))
    ax4.set_yticklabels(sensor_labels, fontsize=8)
    ax4.set_xlabel('Volatility (rate of change std)')
    ax4.set_title('Sensor Volatility Comparison')
    ax4.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.savefig('02_detailed_sensor_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# PRE-EVENT ANALYSIS (24 hours before first cycling event)
# ========================================
print("\n" + "="*60)
print("PRE-EVENT ANALYSIS: 24 HOURS BEFORE FIRST CYCLING EVENT")
print("Target: 24 hours before June 4, 2024 14:00")
print("="*60)

# Define 24-hour window before first cycling event
pre_event_start = first_cycling_event - timedelta(hours=24)
pre_event_end = first_cycling_event

df_pre_event = df_sensors[
    (df_sensors['timestamp'] >= pre_event_start) & 
    (df_sensors['timestamp'] <= pre_event_end)
].copy()

print(f"Pre-event period: {pre_event_start} to {pre_event_end}")
print(f"Sensor records in pre-event window: {len(df_pre_event):,}")

# Analyze each sensor type in the 24-hour pre-event window
sensor_categories = {
    'Demand': demand_sensors,
    'Stage Commands': stage_sensors,
    'Setpoints': setpoint_sensors,
    'Temperature': temp_sensors
}

pre_event_results = {}

for category, sensors in sensor_categories.items():
    print(f"\n### {category.upper()} - 24 Hours Before Cycling")
    print("-" * 50)
    
    category_results = {}
    for sensor in sensors:
        result = analyze_sensor_pattern(df_pre_event, sensor, "24h Pre-Event")
        if result:
            sensor_data, stats = result
            category_results[sensor] = {'data': sensor_data, 'stats': stats}
            
            print(f"\n{sensor} (24h before):")
            print(f"  • Readings: {stats['count']}")
            print(f"  • Mean: {stats['mean']:.2f}")
            print(f"  • Std: {stats['std']:.2f}")
            if 'state_changes' in stats:
                print(f"  • State changes: {stats['state_changes']}")
            if 'volatility' in stats:
                print(f"  • Volatility: {stats['volatility']:.4f}")
    
    pre_event_results[category] = category_results

# ========================================
# VISUALIZATION 3: PRE-EVENT ANALYSIS
# ========================================

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('24-Hour Pre-Event Analysis (June 3-4, 2024)', fontsize=16, fontweight='bold')

# Plot 1: Cooling Demand 24h before
if 'Demand' in pre_event_results and 'Cooling Demand' in pre_event_results['Demand']:
    pre_demand_data = pre_event_results['Demand']['Cooling Demand']['data']
    ax1.plot(pre_demand_data['timestamp'], pre_demand_data['value_numeric'], 
             color='blue', linewidth=2, label='Cooling Demand')
    ax1.axvline(x=first_cycling_event, color='red', linestyle='--', linewidth=2, 
                label='First Cycling Event')
    ax1.set_ylabel('Cooling Demand (%)')
    ax1.set_title('Cooling Demand: 24 Hours Before Cycling')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

# Plot 2: Temperature trends 24h before
temp_colors = ['orange', 'blue', 'green']
for i, (temp_sensor, color) in enumerate(zip(temp_sensors, temp_colors)):
    if 'Temperature' in pre_event_results and temp_sensor in pre_event_results['Temperature']:
        temp_data = pre_event_results['Temperature'][temp_sensor]['data']
        ax2.plot(temp_data['timestamp'], temp_data['value_numeric'], 
                 color=color, alpha=0.7, label=temp_sensor.replace(' Temperature', ''), linewidth=1)

ax2.axvline(x=first_cycling_event, color='red', linestyle='--', linewidth=2)
ax2.set_ylabel('Temperature (°F)')
ax2.set_title('Temperature Trends: 24 Hours Before Cycling')
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Setpoint stability 24h before
for sensor_name in ['Occupied Cool Setpoint', 'Occupied Heat Setpoint']:
    if 'Setpoints' in pre_event_results and sensor_name in pre_event_results['Setpoints']:
        setpoint_data = pre_event_results['Setpoints'][sensor_name]['data']
        ax3.plot(setpoint_data['timestamp'], setpoint_data['value_numeric'], 
                 alpha=0.7, label=sensor_name.replace('Occupied ', ''), linewidth=1)

ax3.axvline(x=first_cycling_event, color='red', linestyle='--', linewidth=2)
ax3.set_ylabel('Setpoint (°F)')
ax3.set_title('Setpoint Stability: 24 Hours Before Cycling')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Volatility comparison 24h vs monthly
comparison_data = []
sensor_names_short = []
monthly_volatility = []
pre_event_volatility = []

for category in ['Demand', 'Setpoints', 'Temperature']:
    if category in pre_event_results:
        for sensor, data in pre_event_results[category].items():
            if sensor in demand_analysis or sensor in setpoint_analysis or sensor in temp_analysis:
                # Get monthly volatility
                if sensor in demand_analysis:
                    monthly_vol = demand_analysis[sensor]['stats'].get('volatility', 0)
                elif sensor in setpoint_analysis:
                    monthly_vol = setpoint_analysis[sensor]['stats'].get('volatility', 0)
                else:
                    monthly_vol = temp_analysis[sensor]['stats'].get('volatility', 0)
                
                pre_vol = data['stats'].get('volatility', 0)
                
                if monthly_vol > 0 and pre_vol > 0:
                    monthly_volatility.append(monthly_vol)
                    pre_event_volatility.append(pre_vol)
                    sensor_names_short.append(sensor.replace(' ', '\n'))

if monthly_volatility:
    x = np.arange(len(sensor_names_short))
    width = 0.35
    
    ax4.bar(x - width/2, monthly_volatility, width, label='Monthly Average', alpha=0.7)
    ax4.bar(x + width/2, pre_event_volatility, width, label='24h Pre-Event', alpha=0.7)
    
    ax4.set_ylabel('Volatility')
    ax4.set_title('Volatility: Monthly vs 24h Pre-Event')
    ax4.set_xticks(x)
    ax4.set_xticklabels(sensor_names_short, fontsize=8, rotation=45)
    ax4.legend()
    ax4.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('03_pre_event_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# COMPARATIVE ANALYSIS: NORMAL vs PRE-CYCLING
# ========================================
print("\n" + "="*60)
print("COMPARATIVE ANALYSIS: NORMAL vs PRE-CYCLING PERIODS")
print("="*60)

# Define "normal" period (before any cycling issues)
normal_start = datetime(2024, 5, 1)
normal_end = datetime(2024, 5, 31)

df_normal = df_sensors[
    (df_sensors['timestamp'] >= normal_start) & 
    (df_sensors['timestamp'] <= normal_end)
].copy()

print(f"Normal period: {normal_start.date()} to {normal_end.date()}")
print(f"Pre-cycling period: {pre_cycling_start.date()} to {pre_cycling_end.date()}")

def compare_periods(sensor_name, normal_df, precycling_df):
    """Compare sensor behavior between normal and pre-cycling periods"""
    
    normal_result = analyze_sensor_pattern(normal_df, sensor_name, "Normal")
    precycling_result = analyze_sensor_pattern(precycling_df, sensor_name, "Pre-Cycling")
    
    if not normal_result or not precycling_result:
        return None
    
    normal_data, normal_stats = normal_result
    precycling_data, precycling_stats = precycling_result
    
    # Calculate differences
    comparison = {
        'sensor': sensor_name,
        'normal_mean': normal_stats['mean'],
        'precycling_mean': precycling_stats['mean'],
        'mean_change': precycling_stats['mean'] - normal_stats['mean'],
        'normal_std': normal_stats['std'],
        'precycling_std': precycling_stats['std'],
        'std_change': precycling_stats['std'] - normal_stats['std'],
        'normal_range': normal_stats['range'],
        'precycling_range': precycling_stats['range'],
        'range_change': precycling_stats['range'] - normal_stats['range']
    }
    
    if 'volatility' in normal_stats and 'volatility' in precycling_stats:
        comparison['normal_volatility'] = normal_stats['volatility']
        comparison['precycling_volatility'] = precycling_stats['volatility']
        comparison['volatility_change'] = precycling_stats['volatility'] - normal_stats['volatility']
    
    if 'cycling_frequency' in normal_stats and 'cycling_frequency' in precycling_stats:
        comparison['normal_cycling'] = normal_stats['cycling_frequency']
        comparison['precycling_cycling'] = precycling_stats['cycling_frequency']
        comparison['cycling_change'] = precycling_stats['cycling_frequency'] - normal_stats['cycling_frequency']
    
    return comparison

# Perform comparisons for key sensors
print("\n### PERIOD COMPARISON RESULTS")
print("-" * 40)

significant_changes = []
comparison_results = []

for sensor in cooling_sensors[:10]:  # Analyze top 10 cooling sensors
    comparison = compare_periods(sensor, df_normal, df_period)
    if comparison:
        comparison_results.append(comparison)
        print(f"\n{sensor}:")
        print(f"  • Mean: {comparison['normal_mean']:.2f} → {comparison['precycling_mean']:.2f} (Δ{comparison['mean_change']:+.2f})")
        print(f"  • Std:  {comparison['normal_std']:.2f} → {comparison['precycling_std']:.2f} (Δ{comparison['std_change']:+.2f})")
        print(f"  • Range: {comparison['normal_range']:.2f} → {comparison['precycling_range']:.2f} (Δ{comparison['range_change']:+.2f})")
        
        # Check for significant changes
        if abs(comparison['std_change']) > comparison['normal_std'] * 0.5:  # 50% increase in volatility
            significant_changes.append(f"{sensor}: Volatility increased by {comparison['std_change']:.2f}")
        
        if 'volatility_change' in comparison and comparison['volatility_change'] > 0:
            print(f"  • Volatility: {comparison['normal_volatility']:.4f} → {comparison['precycling_volatility']:.4f} (Δ{comparison['volatility_change']:+.4f})")
        
        if 'cycling_change' in comparison:
            print(f"  • Cycling Freq: {comparison['normal_cycling']:.2f} → {comparison['precycling_cycling']:.2f} (Δ{comparison['cycling_change']:+.2f})")
            if comparison['cycling_change'] > 1.0:  # Significant increase in cycling
                significant_changes.append(f"{sensor}: Cycling frequency increased by {comparison['cycling_change']:.2f} changes/hour")

# ========================================
# VISUALIZATION 4: COMPARATIVE ANALYSIS
# ========================================

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Comparative Analysis: Normal (May) vs Pre-Cycling (June) Periods', fontsize=16, fontweight='bold')

# Plot 1: Mean value changes
if comparison_results:
    sensors = [comp['sensor'][:15] + '...' if len(comp['sensor']) > 15 else comp['sensor'] for comp in comparison_results]
    mean_changes = [comp['mean_change'] for comp in comparison_results]
    
    colors = ['red' if x > 0 else 'blue' for x in mean_changes]
    ax1.barh(range(len(sensors)), mean_changes, color=colors, alpha=0.7)
    ax1.set_yticks(range(len(sensors)))
    ax1.set_yticklabels(sensors, fontsize=8)
    ax1.set_xlabel('Mean Value Change')
    ax1.set_title('Mean Value Changes: May → June')
    ax1.axvline(x=0, color='black', linestyle='-', alpha=0.3)
    ax1.grid(axis='x', alpha=0.3)

# Plot 2: Standard deviation changes
if comparison_results:
    std_changes = [comp['std_change'] for comp in comparison_results]
    colors = ['red' if x > 0 else 'blue' for x in std_changes]
    ax2.barh(range(len(sensors)), std_changes, color=colors, alpha=0.7)
    ax2.set_yticks(range(len(sensors)))
    ax2.set_yticklabels(sensors, fontsize=8)
    ax2.set_xlabel('Standard Deviation Change')
    ax2.set_title('Volatility Changes: May → June')
    ax2.axvline(x=0, color='black', linestyle='-', alpha=0.3)
    ax2.grid(axis='x', alpha=0.3)

# Plot 3: Before/After comparison for key sensors (with proper unit scaling)
periods = ['Normal (May)', 'Pre-Cycling (June)']

if comparison_results:
    # Create twin axis for different units
    ax3_temp = ax3.twinx()
    
    # Cooling Demand on left axis (%)
    demand_comp = next((c for c in comparison_results if c['sensor'] == 'Cooling Demand'), None)
    if demand_comp:
        x_pos = np.array([0, 1]) - 0.2
        values = [demand_comp['normal_mean'], demand_comp['precycling_mean']]
        stds = [demand_comp['normal_std'], demand_comp['precycling_std']]
        ax3.bar(x_pos, values, 0.3, yerr=stds, alpha=0.7, color='pink', 
               label='Cooling Demand (%)', capsize=3)
    
    # Temperature sensors on right axis (°F)
    temp_sensors = ['Outside Temperature', 'Occupied Cool Setpoint']
    temp_colors = ['orange', 'green']
    
    for i, (sensor, color) in enumerate(zip(temp_sensors, temp_colors)):
        comp = next((c for c in comparison_results if c['sensor'] == sensor), None)
        if comp:
            x_pos = np.array([0, 1]) + i * 0.3
            values = [comp['normal_mean'], comp['precycling_mean']]
            stds = [comp['normal_std'], comp['precycling_std']]
            ax3_temp.bar(x_pos, values, 0.3, yerr=stds, alpha=0.7, color=color,
                        label=f'{sensor} (°F)', capsize=3)
    
    # Formatting
    ax3.set_xlabel('Period')
    ax3.set_ylabel('Cooling Demand (%)', color='pink')
    ax3_temp.set_ylabel('Temperature (°F)', color='blue')
    ax3.set_title('Key Sensor Comparison: Normal vs Pre-Cycling')
    ax3.set_xticks([0, 1])
    ax3.set_xticklabels(periods)
    
    # Combine legends
    lines1, labels1 = ax3.get_legend_handles_labels()
    lines2, labels2 = ax3_temp.get_legend_handles_labels()
    ax3.legend(lines1 + lines2, labels1 + labels2, fontsize=8, loc='upper left')
    
    ax3.grid(axis='y', alpha=0.3)

# Plot 4: Volatility scatter plot
if comparison_results:
    normal_volatilities = []
    precycling_volatilities = []
    sensor_labels = []
    
    for comp in comparison_results:
        if 'normal_volatility' in comp and 'precycling_volatility' in comp:
            normal_volatilities.append(comp['normal_volatility'])
            precycling_volatilities.append(comp['precycling_volatility'])
            sensor_labels.append(comp['sensor'][:10] + '...' if len(comp['sensor']) > 10 else comp['sensor'])
    
    if normal_volatilities:
        ax4.scatter(normal_volatilities, precycling_volatilities, alpha=0.7, s=100)
        
        # Add diagonal line (y=x)
        max_val = max(max(normal_volatilities), max(precycling_volatilities))
        ax4.plot([0, max_val], [0, max_val], 'r--', alpha=0.5, label='No Change Line')
        
        # Label points
        for i, label in enumerate(sensor_labels):
            ax4.annotate(label, (normal_volatilities[i], precycling_volatilities[i]), 
                        xytext=(5, 5), textcoords='offset points', fontsize=8)
        
        ax4.set_xlabel('Normal Period Volatility')
        ax4.set_ylabel('Pre-Cycling Period Volatility')
        ax4.set_title('Volatility: Normal vs Pre-Cycling')
        ax4.legend()
        ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('04_comparative_analysis.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# VISUALIZATION 5: COOLING DEMAND DEEP DIVE
# ========================================

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Cooling Demand Deep Dive Analysis', fontsize=16, fontweight='bold')

# Get cooling demand data for both periods
normal_demand = df_normal[df_normal['navName'] == 'Cooling Demand'].copy()
normal_demand['value_numeric'] = pd.to_numeric(normal_demand['value'], errors='coerce')
normal_demand = normal_demand.dropna().sort_values('timestamp')

precycling_demand = df_period[df_period['navName'] == 'Cooling Demand'].copy()
precycling_demand['value_numeric'] = pd.to_numeric(precycling_demand['value'], errors='coerce')
precycling_demand = precycling_demand.dropna().sort_values('timestamp')

# Plot 1: Time series comparison
if len(normal_demand) > 0:
    ax1.plot(normal_demand['timestamp'], normal_demand['value_numeric'], 
             alpha=0.7, label='Normal (May)', color='blue', linewidth=1)

if len(precycling_demand) > 0:
    ax1.plot(precycling_demand['timestamp'], precycling_demand['value_numeric'], 
             alpha=0.7, label='Pre-Cycling (June)', color='red', linewidth=1)
    
    # Mark cycling events
    for event in cycling_events:
        ax1.axvline(x=event, color='orange', linestyle='--', alpha=0.7)

ax1.set_ylabel('Cooling Demand (%)')
ax1.set_title('Cooling Demand: Normal vs Pre-Cycling Periods')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot 2: Distribution comparison
if len(normal_demand) > 0 and len(precycling_demand) > 0:
    ax2.hist(normal_demand['value_numeric'], bins=20, alpha=0.7, 
             label='Normal (May)', color='blue', density=True)
    ax2.hist(precycling_demand['value_numeric'], bins=20, alpha=0.7, 
             label='Pre-Cycling (June)', color='red', density=True)
    
    ax2.set_xlabel('Cooling Demand (%)')
    ax2.set_ylabel('Density')
    ax2.set_title('Cooling Demand Distribution Comparison')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

# Plot 3: Daily patterns
if len(precycling_demand) > 0:
    precycling_demand['hour'] = precycling_demand['timestamp'].dt.hour
    hourly_avg = precycling_demand.groupby('hour')['value_numeric'].mean()
    hourly_std = precycling_demand.groupby('hour')['value_numeric'].std()
    
    ax3.plot(hourly_avg.index, hourly_avg.values, 'o-', color='red', alpha=0.7)
    ax3.fill_between(hourly_avg.index, 
                     hourly_avg.values - hourly_std.values,
                     hourly_avg.values + hourly_std.values,
                     alpha=0.3, color='red')
    
    # Mark typical cycling hours
    cycling_hours = [event.hour for event in cycling_events]
    for hour in set(cycling_hours):
        ax3.axvline(x=hour, color='orange', linestyle='--', alpha=0.7)
    
    ax3.set_xlabel('Hour of Day')
    ax3.set_ylabel('Average Cooling Demand (%)')
    ax3.set_title('Daily Cooling Demand Pattern (June)')
    ax3.grid(True, alpha=0.3)
    ax3.set_xticks(range(0, 24, 3))

# Plot 4: Rate of change analysis
if len(precycling_demand) > 1:
    precycling_demand['time_diff'] = precycling_demand['timestamp'].diff().dt.total_seconds() / 60
    precycling_demand['value_diff'] = precycling_demand['value_numeric'].diff()
    precycling_demand['rate_of_change'] = precycling_demand['value_diff'] / precycling_demand['time_diff']
    
    # Remove outliers for better visualization
    rate_data = precycling_demand['rate_of_change'].dropna()
    q1, q3 = rate_data.quantile([0.25, 0.75])
    iqr = q3 - q1
    filtered_rates = rate_data[(rate_data >= q1 - 1.5*iqr) & (rate_data <= q3 + 1.5*iqr)]
    
    ax4.hist(filtered_rates, bins=30, alpha=0.7, color='red', density=True)
    ax4.axvline(x=0, color='black', linestyle='-', alpha=0.5)
    ax4.set_xlabel('Rate of Change (%/min)')
    ax4.set_ylabel('Density')
    ax4.set_title('Cooling Demand Rate of Change (June)')
    ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('05_cooling_deep_dive.png', dpi=300, bbox_inches='tight')
plt.show()

# ========================================
# EARLY WARNING SIGNALS DETECTION
# ========================================
print("\n" + "="*60)
print("EARLY WARNING SIGNALS DETECTION")
print("="*60)

print("\n### SIGNIFICANT CHANGES DETECTED:")
if significant_changes:
    for change in significant_changes:
        print(f"  ⚠️  {change}")
else:
    print("  • No significant changes detected with current thresholds")

print("\n### POTENTIAL EARLY WARNING INDICATORS:")
print("Based on analysis, look for these patterns before cycling events:")

# Summarize key findings for early warning system
warning_indicators = []

# Check for sensors with increased volatility
for category, analysis_dict in [
    ('Demand', demand_analysis),
    ('Stage Commands', stage_analysis), 
    ('Setpoints', setpoint_analysis),
    ('Temperature', temp_analysis)
]:
    for sensor, data in analysis_dict.items():
        stats = data['stats']
        if 'volatility' in stats and stats['volatility'] > 0:
            warning_indicators.append(f"• {sensor}: Monitor for volatility > {stats['volatility']:.4f}")

print("\n".join(warning_indicators[:5]))  # Top 5 indicators

# ========================================
# VISUALIZATION 6: EARLY WARNING DASHBOARD
# ========================================

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Early Warning Dashboard - Key Indicators', fontsize=16, fontweight='bold')

# Plot 1: Warning indicator thresholds
if warning_indicators:
    indicator_names = []
    threshold_values = []
    
    for indicator in warning_indicators[:6]:  # Top 6 indicators
        parts = indicator.split(': Monitor for volatility > ')
        if len(parts) == 2:
            name = parts[0].replace('• ', '')
            threshold = float(parts[1])
            indicator_names.append(name.replace(' ', '\n'))
            threshold_values.append(threshold)
    
    if threshold_values:
        colors = plt.cm.Reds(np.linspace(0.3, 1, len(threshold_values)))
        ax1.barh(range(len(indicator_names)), threshold_values, color=colors)
        ax1.set_yticks(range(len(indicator_names)))
        ax1.set_yticklabels(indicator_names, fontsize=8)
        ax1.set_xlabel('Volatility Threshold')
        ax1.set_title('Early Warning Thresholds')
        ax1.grid(axis='x', alpha=0.3)

# Plot 2: Timeline of key events and patterns
timeline_events = []
timeline_labels = []

# Add cycling events
for i, event in enumerate(cycling_events):
    timeline_events.append(event)
    timeline_labels.append(f'Cycling Event {i+1}')

# Add analysis periods
timeline_events.extend([
    datetime(2024, 5, 1),   # Normal period start
    datetime(2024, 6, 1),   # Pre-cycling period start
    pre_event_start,        # 24h pre-event start
])
timeline_labels.extend([
    'Normal Period Start',
    'Pre-Cycling Period Start',
    '24h Pre-Event Window'
])

# Sort by date
timeline_data = sorted(zip(timeline_events, timeline_labels))
timeline_events, timeline_labels = zip(*timeline_data)

colors = ['red' if 'Cycling' in label else 'blue' if 'Normal' in label else 'orange' for label in timeline_labels]
ax2.scatter(timeline_events, range(len(timeline_events)), c=colors, s=100, alpha=0.7)

for i, (event, label) in enumerate(zip(timeline_events, timeline_labels)):
    ax2.annotate(label, (event, i), xytext=(10, 0), textcoords='offset points', 
                fontsize=8, ha='left')

ax2.set_ylabel('Event Sequence')
ax2.set_title('Analysis Timeline and Key Events')
ax2.grid(True, alpha=0.3)

# Plot 3: Risk assessment matrix
risk_sensors = ['Cooling Demand', 'Occupied Cool Setpoint', 'Occupied Heat Setpoint', 'Discharge Temperature']
risk_levels = []
impact_levels = []
sensor_names_clean = []

for sensor in risk_sensors:
    comp = next((c for c in comparison_results if c['sensor'] == sensor), None)
    if comp:
        # Calculate risk level based on volatility change
        if 'volatility_change' in comp:
            risk_level = min(comp['volatility_change'] * 10, 5)  # Scale to 0-5
        else:
            risk_level = abs(comp['std_change']) / comp['normal_std'] if comp['normal_std'] > 0 else 0
        
        # Calculate impact based on mean change
        impact_level = abs(comp['mean_change']) / comp['normal_mean'] if comp['normal_mean'] > 0 else 0
        impact_level = min(impact_level * 5, 5)  # Scale to 0-5
        
        risk_levels.append(risk_level)
        impact_levels.append(impact_level)
        sensor_names_clean.append(sensor.replace(' ', '\n'))

if risk_levels:
    scatter = ax3.scatter(risk_levels, impact_levels, s=200, alpha=0.7, c=range(len(risk_levels)), cmap='viridis')
    
    for i, name in enumerate(sensor_names_clean):
        ax3.annotate(name, (risk_levels[i], impact_levels[i]), 
                    xytext=(5, 5), textcoords='offset points', fontsize=8)
    
    ax3.set_xlabel('Risk Level (Volatility Change)')
    ax3.set_ylabel('Impact Level (Mean Change)')
    ax3.set_title('Risk Assessment Matrix')
    ax3.grid(True, alpha=0.3)
    
    # Add quadrant lines
    ax3.axhline(y=2.5, color='red', linestyle='--', alpha=0.5)
    ax3.axvline(x=2.5, color='red', linestyle='--', alpha=0.5)

# Plot 4: Predictive confidence levels
confidence_data = {
    'Cooling Demand Spike': 85,
    'Setpoint Hunting': 75,
    'Temperature Instability': 70,
    'Equipment Cycling': 80,
    'Seasonal Transition': 90
}

indicators = list(confidence_data.keys())
confidences = list(confidence_data.values())
colors = ['red' if conf >= 80 else 'orange' if conf >= 70 else 'yellow' for conf in confidences]

ax4.barh(indicators, confidences, color=colors, alpha=0.7)
ax4.set_xlabel('Confidence Level (%)')
ax4.set_title('Early Warning Indicator Confidence')
ax4.grid(axis='x', alpha=0.3)

# Add confidence threshold line
ax4.axvline(x=75, color='red', linestyle='--', alpha=0.5, label='High Confidence Threshold')
ax4.legend()

plt.tight_layout()
plt.savefig('06_early_warning_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print(f"\n### NEXT STEPS:")
print("1. Validate these patterns against the other cycling events (June 17, 21, 26)")
print("2. Create automated thresholds based on normal vs pre-cycling patterns")
print("3. Develop real-time monitoring for the identified warning indicators")
print("4. Test early warning system on May 2025 cycling event")

print("\n" + "="*60)
print("PHASE 1 ANALYSIS COMPLETE")
print("="*60)

# ========================================
# SUMMARY STATISTICS FOR PRESENTATION
# ========================================

print(f"\n### KEY FINDINGS SUMMARY FOR STAKEHOLDER PRESENTATION:")
print("-" * 60)

print(" DATA OVERVIEW:")
print(f"  • Analysis Period: June 2024 (pre-cycling detection)")
print(f"  • Total Sensor Records: {len(df_period):,}")
print(f"  • Cooling-Related Sensors: {len(cooling_sensors)}")
print(f"  • Cycling Events Analyzed: {len(cycling_events)}")

print("\n CRITICAL FINDINGS:")
if 'Cooling Demand' in [comp['sensor'] for comp in comparison_results]:
    cooling_comp = next(comp for comp in comparison_results if comp['sensor'] == 'Cooling Demand')
    mean_increase = (cooling_comp['mean_change'] / cooling_comp['normal_mean']) * 100
    std_increase = (cooling_comp['std_change'] / cooling_comp['normal_std']) * 100
    print(f"  • Cooling Demand increased {mean_increase:.1f}% before cycling events")
    print(f"  • Cooling Demand volatility increased {std_increase:.1f}%")

print(f"  • {len(significant_changes)} sensors showed significant pattern changes")
print(f"  • Early warning possible {(first_cycling_event - pre_cycling_start).days} days before first event")

print("\n TIMING INSIGHTS:")
cycling_hours = [event.hour for event in cycling_events]
print(f"  • Cycling events occurred at hours: {sorted(set(cycling_hours))}")
print(f"  • Peak cycling risk: {max(set(cycling_hours), key=cycling_hours.count)}:00")

print("\n ACTIONABLE RECOMMENDATIONS:")
print("  • Implement cooling demand monitoring when outdoor temp > 70°F")
print("  • Set volatility alerts for cooling demand > 1.54")
print("  • Monitor setpoint stability during seasonal transitions")
print("  • Establish 24-hour pre-event monitoring protocols")
```
