```python
# ------------------------------------------------------------
# 0) Imports  – add SciPy for the MAD outlier filter
# ------------------------------------------------------------
%pip install scipy --quiet

import numpy as np
from scipy.stats import median_abs_deviation


# ------------------------------------------------------------
# 1) One-time column cleanup  (run once, right after the merge)
# ------------------------------------------------------------
cols_to_drop = ["Unnamed: 0",               # helper index – clutters df   [oai_citation:0‡data_analysis.pdf](file-service://file-AEEQF1gdyqCdtm9E1rLpXB)
                "equipRef_x", "equipRef_y", # duplicate meta
                "unit_y"]                   # we keep the X version
merged_df = (
    merged_df
      .drop(columns=[c for c in cols_to_drop if c in merged_df])
      .rename(columns={"unit_x": "unit"})    # keep a single, clear “unit” col
)


# ------------------------------------------------------------
# 2) Type casting & smarter timestamp handling
# ------------------------------------------------------------
#   – Convert once, BEFORE splitting by unit
merged_df["value"]     = pd.to_numeric(merged_df["value"], errors="coerce")
merged_df["timestamp"] = (
    pd.to_datetime(merged_df["timestamp"])          # parse
      .dt.tz_localize("America/New_York",           # source tz comes from tz col   [oai_citation:1‡data_analysis.pdf](file-service://file-AEEQF1gdyqCdtm9E1rLpXB)
                      ambiguous="NaT", nonexistent="shift_forward")
)


# Set index + sort for fast time-group ops
merged_df = merged_df.set_index("timestamp").sort_index()


# ------------------------------------------------------------
# 3) Robust MAD-based outlier removal on the fly
# ------------------------------------------------------------
def mad_filter(series, k=5):
    """Return series with k*MAD outliers set to NaN (robust to spikes)."""
    med = np.nanmedian(series)
    mad = median_abs_deviation(series, nan_policy="omit")
    print("Median:", med)
    print("MAD:", mad)
    return series.where(np.abs(series - med) <= k * mad)


# Apply only to numeric °F points (e.g., Discharge Temp, Zone Temp)
is_temp = merged_df["unit"].eq("°F") & merged_df["value"].notna()
merged_df.loc[is_temp, "value"] = mad_filter(merged_df.loc[is_temp, "value"])


# ------------------------------------------------------------
# 4) Build an **hourly** wide table for all numeric signals
#    [oai_citation:2‡data_analysis.pdf](file-service://file-AEEQF1gdyqCdtm9E1rLpXB)
# ------------------------------------------------------------
numeric_df = merged_df[merged_df["kind"].eq("Number")]


wide_h = (
    numeric_df
      .pivot_table(index=pd.Grouper(freq="h"),   # <- lower-case
                   columns="navName",
                   values="value",
                   aggfunc="mean")
)


# ------------------------------------------------------------
# 5) Quick seasonal decomposition demo on Cooling Demand
# ------------------------------------------------------------
%pip install statsmodels --quiet

from statsmodels.tsa.seasonal import STL


if "Cooling Demand" in wide_h.columns:
    cd = wide_h["Cooling Demand"].dropna()
    stl = STL(cd, period=24*7).fit()   # weekly seasonality
    stl.plot()
    # make the plot larger
    plt.gcf().set_size_inches(12, 8)



```

### Exploratory questions 

1. **Does outdoor air temperature drive Cooling-Demand in our data at all?**  
2. **If yes, is the response immediate or delayed?**&nbsp;→ *find the lag*  
3. **What is the typical delay (hours) where the link is strongest?**  
4. **How strong is that link at each lag (Pearson *r*)?**  
5. **After the best-fit delay, roughly how much extra load does each +1 °F add?**  
6. **Is this lag-and-sensitivity relationship stable across the whole cooling season (Apr – Oct)?**  






### Residual Analysis & Fault Flagging

We already decomposed Cooling Demand into Trend + Seasonal + Residual.


Under normal operation the residuals should hover around 0.


Hours where the residual shoots far above (or below) the typical spread often coincide with a fault.


So we flag large residuals as “events worth a look".




```python
# ------------------------------------------------------------------
# 0.  Imports
# ------------------------------------------------------------------
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from  scipy.signal import correlate
import statsmodels.formula.api as smf
from  statsmodels.tsa.seasonal import STL

# ------------------------------------------------------------------
# 1.  Completeness check
# ------------------------------------------------------------------
pair = wide_h[['Cooling Demand', 'Outside Temperature']]

good_rows = pair.dropna().shape[0]
print(f"{good_rows:,}/{len(pair):,} rows "
      f"({100*good_rows/len(pair):.1f} %) have BOTH series present")

# (pair.isna()
#      .astype(int)
#      .rolling(24).sum()
#      .plot(title='Rolling 24-h count of NaNs',
#            xlabel='Timestamp', ylabel='Missing-value count'));

# ------------------------------------------------------------------
# 2.  Fill ≤30 h gaps in OA temperature
# ------------------------------------------------------------------
pair_filled = pair.copy()
pair_filled['Outside Temperature'] = (
    pair_filled['Outside Temperature']
        .interpolate(limit=30, limit_direction='both')
)

pair_clean = pair_filled.dropna()
print(f"Data kept after fill: {len(pair_clean):,} "
      f"({100*len(pair_clean)/len(pair):.1f} %)")

# fig, ax = plt.subplots(figsize=(6, 3))
# wide_h['Outside Temperature'].plot(ax=ax, label='raw', alpha=.3)
# pair_clean['Outside Temperature'].plot(ax=ax, label='filled', lw=1)
# ax.set_title('Outside Temperature: raw vs filled')
# ax.set_ylabel('°F'); ax.legend(); fig.tight_layout(); plt.show()

# ------------------------------------------------------------------
# 3.  Cross-correlation
# ------------------------------------------------------------------
def ccf(a, b, max_lag=24):
    a = (a - a.mean())/a.std();  b = (b - b.mean())/b.std()
    corr = correlate(a, b, mode='full')/len(a)
    lags = np.arange(-len(a)+1, len(a))
    idx  = (lags >= -max_lag) & (lags <= max_lag)
    return lags[idx], corr[idx]

lags, corr = ccf(pair_clean['Outside Temperature'],
                 pair_clean['Cooling Demand'])
best_lag = int(lags[np.argmax(corr)])
print(f"Best lag = {best_lag:+} h   (ρ = {corr.max():.2f})")

# ------------------------------------------------------------------
# 4.  Simple linear model with lagged temperature
# ------------------------------------------------------------------
wide_h['T_out_lead'] = wide_h['Outside Temperature'].shift(-best_lag)
df_model = wide_h[['Cooling Demand', 'T_out_lead']].dropna()

model = smf.ols("Q('Cooling Demand') ~ Q('T_out_lead')",
                data=df_model).fit()
print(model.summary())

# ------------------------------------------------------------------
# 5.  STL residuals & histogram
# ------------------------------------------------------------------
resid = (df_model['Cooling Demand'] - model.fittedvalues)
stl   = STL(resid, period=24*7).fit()
resid = stl.resid.dropna()

# Robust z-score & anomalies (optional)
med = resid.median();  mad = (abs(resid - med)).median()
zrob = (resid - med)/(1.4826*mad)
print(f"Anomalies (|z|>5): {(abs(zrob)>5).sum()} hours")

# Clean histogram
fig, ax = plt.subplots(figsize=(6, 3))
ax.hist(resid.values, bins=50, edgecolor='black')
ax.set_title('Residuals after 4-h lag correction')
ax.set_xlabel('Residual (%)')          # ← x-axis label
ax.set_ylabel('Hour count')             # ← y-axis label
fig.tight_layout(); plt.show()
```

* **Slope ≈ 0.52 % per °F** – every 1 °F rise in outdoor air (4 h earlier) drives the controller to request ~0.5 % more cooling.  
* **R² ≈ 0.24** – outdoor air alone explains about a quarter of the demand signal; occupancy and set-points make up the rest.  
* **224 extreme hours (|z| > 5)** – mostly positive residuals, highlighting periods when the plant requests far more cooling than the weather alone would predict.

The residual histogram centres on zero and is bell-shaped, confirming the 4-hour shift removed bias. The long right tail captures those over-cooling events we’ll flag for fault review.

```python
# =================================================================
# Bulk-normalise sensors and cross-correlate vs. a target variable
# =================================================================
import pandas as pd, numpy as np
from pathlib import Path
from scipy.signal import correlate
from typing import Union

# ------------------------------------------------------------
# 0. Helper functions
# ------------------------------------------------------------
def load_sensor_csv(path: Union[str, Path]) -> pd.DataFrame:
    """Return DF with ['timestamp','navName','value'] – auto-detects headers."""
    df = pd.read_csv(path, low_memory=False)        # <- no dtype warning

    # --- find timestamp column ------------------------------------------------
    time_col = next((c for c in df.columns
                     if c.lower().startswith(('time', 'date'))), None)
    if time_col is None:
        raise ValueError(f"No timestamp column found in {path}")

    # --- standardise names ----------------------------------------------------
    col_map = {time_col: 'timestamp'}
    for cand in ['navName', 'NavName', 'sensor', 'point']:
        if cand in df.columns:
            col_map[cand] = 'navName'; break
    for cand in ['value', 'Value', 'reading']:
        if cand in df.columns:
            col_map[cand] = 'value'; break
    df = df.rename(columns=col_map)

    missing = {'timestamp', 'navName', 'value'} - set(df.columns)
    if missing:
        raise ValueError(f"{path}: missing columns {missing}")

    df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
    return df[['timestamp', 'navName', 'value']]

def pivot_hourly(df: pd.DataFrame, how: str = 'mean') -> pd.DataFrame:
    """Return wide hourly table from long sensor DF."""
    df = df.set_index('timestamp')
    if how == 'mean':
        df['value'] = pd.to_numeric(df['value'], errors='coerce')
        agg = df.groupby('navName')['value'].resample('h').mean()   # <-- 'h'
    else:
        agg = df.groupby('navName')['value'].resample('h').first()
    return agg.unstack(level=0)

def zscore(s: pd.Series) -> pd.Series:
    return (s - s.mean()) / s.std()

def ccf_fast(a, b, max_lag: int = 24):
    corr = correlate(a, b, mode='full') / len(a)
    lags = np.arange(-len(a)+1, len(a))
    mask = (lags >= -max_lag) & (lags <= max_lag)
    lags, corr = lags[mask], corr[mask]
    k = np.argmax(np.abs(corr))
    return int(lags[k]), corr[k]

# ------------------------------------------------------------
# 1. Load merged file (has timestamp)
# ------------------------------------------------------------
MERGED_PATH = "comerica_rtu_merged_data.csv"          # adjust if needed
merged = pd.read_csv(MERGED_PATH, low_memory=False)
if 'timestamp' not in merged.columns:
    raise ValueError("merged file must contain a 'timestamp' column")
merged['timestamp'] = pd.to_datetime(merged['timestamp'], errors='coerce')

# Strip trailing '%' and coerce to float so % sensors are numeric
unit_col = None
for col in ['unit', 'unit_x', 'unit_y']:
    if col in merged.columns:
        unit_col = col
        break
if unit_col is None:
    raise KeyError("No 'unit', 'unit_x', or 'unit_y' column found in merged DataFrame")

pct_mask = merged[unit_col].str.contains('%', na=False)
merged.loc[pct_mask, 'value'] = (
    merged.loc[pct_mask, 'value']
          .astype(str)
          .str.rstrip('%')
          .replace('', np.nan)
          .astype(float)
)

# ------------------------------------------------------------
# 2. Split by unit and write helper CSVs (optional but keeps parity)
# ------------------------------------------------------------
temps = merged[merged[unit_col].str.contains('F', na=False)].copy()
pct   = merged[pct_mask].copy()
nonn  = merged[~merged[unit_col].str.contains('F|%', na=False, regex=True)].copy()

temps.to_csv("comerica_rtu_temperature_data.csv", index=False)
pct.to_csv  ("comerica_rtu_percentage_data.csv", index=False)
nonn.to_csv ("comerica_rtu_non_numeric_data.csv", index=False)
print("✓ Three unit-based files written")

# ------------------------------------------------------------
# 3. Build one hourly wide DataFrame
# ------------------------------------------------------------
wide = pivot_hourly(temps, 'mean')
wide = wide.join(pivot_hourly(pct,  'mean'),  how='outer')
wide = wide.join(pivot_hourly(nonn, 'first'), how='outer')

# ------------------------------------------------------------
# 4. Normalise numeric / boolean columns
# ------------------------------------------------------------
num_cols  = wide.select_dtypes('number').columns
bool_cols = wide.select_dtypes('bool').columns
str_cols  = wide.select_dtypes('object').columns

wide[num_cols]  = wide[num_cols].apply(zscore, axis=0)
wide[bool_cols] = (wide[bool_cols].astype(float)
                              .apply(zscore, axis=0))
wide            = wide.drop(columns=str_cols)     # drop pure strings
wide            = wide.dropna(how='any')          # row alignment

print(f"Aligned dataset: {len(wide):,} hourly rows · {wide.shape[1]} sensors")

# ------------------------------------------------------------
# 5. Cross-correlate vs. target
# ------------------------------------------------------------
TARGET, MAX_LAG = 'Cooling Demand', 24
if TARGET not in wide.columns:
    raise ValueError(f"Target '{TARGET}' not in data")

target = wide[TARGET]
results = []
for col in wide.columns:
    if col == TARGET:
        continue
    lag, rho = ccf_fast(target, wide[col], MAX_LAG)
    results.append((col, lag, rho))

ccf_df = (pd.DataFrame(results, columns=['sensor', 'best_lag_h', 'rho'])
            .sort_values('rho', key=np.abs, ascending=False)
            .reset_index(drop=True))

print("\nTop 20 sensors by |ρ| against", TARGET)
print(ccf_df.head(20).to_string(index=False))

```

Let'a dive into Discharge Temperature, because it’s (a) highly correlated, (b) timed exactly with coil activity, and (c) a direct measure of delivered cooling rather than a command, Discharge Temperature is the most useful sensor to drill into next.




```python
from matplotlib import pyplot as plt

cand = 'Discharge Temperature'
lags = np.arange(-24, 25)
cors = [wide[TARGET].corr(wide[cand].shift(-k)) for k in lags]
plt.plot(lags, cors, marker='o'); plt.axhline(0, ls='--')
plt.title(f'CCF vs {cand}'); plt.xlabel('Lag (h)'); plt.ylabel('ρ');

```

| Feature of the curve                        | What it means                                                                                                 |
|---------------------------------------------|---------------------------------------------------------------------------------------------------------------|
| Deep negative spike at lag 0 (≈ –0.40)      | When discharge-air temperature (DAT) is high in a given hour, Cooling Demand is low that same hour (and vice-versa). That makes sense: colder supply air means the cooling coil is working harder. |
| Positive humps peaking around –10 h and +10 h (≈ +0.10) | Because DAT and Cooling Demand both follow the daily schedule, they drift in and out of phase every ~12 hours, creating those mirror-image lobes. |
| Near-symmetry left vs. right                | Shows the two series are dominated by a 24-hour cycle rather than one leading the other for many hours.        |
| Best absolute correlation at lag 0          | Confirms DAT reacts essentially in the same hour as the cooling load—no meaningful lead or lag between them.   |

**Take-away:** Discharge Temperature is an instantaneous indicator of cooling effort (strong inverse link at lag 0). The side lobes are just the daily rhythm of the HVAC schedule, not a causal delay. That matches expectations, so the result is perfectly reasonable.

```python
# ------------------------------------------------------------
# 2 A – pull residuals & compute a robust z-score
# ------------------------------------------------------------
resid = stl.resid.dropna()                       # residual is already “Cooling Demand % – model”

# Median & MAD
med  = resid.median()
mad  = (np.abs(resid - med)).median()
zrob = (resid - med) / (1.4826 * mad)            # 1.4826 × MAD ≈ 1 σ for a normal dist.

# Flag anomalies: |z| > 5   (adjust if you need a looser cut)
anomaly_mask = zrob.abs() >5
anomalies    = resid[anomaly_mask]

print(f"{anomalies.size} anomalous hours found "
      f"({100*anomalies.size/len(resid):.1f} % of the season)")

# ------------------------------------------------------------
# 2 B – quick visual check
# ------------------------------------------------------------
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(resid.index, resid, label="Residual (%)", lw=0.7)
ax.scatter(anomalies.index, anomalies, color="red", s=18,
           label="Flagged anomaly")
ax.axhline(0, color="black", lw=0.8)
ax.set_title("Cooling-Demand residuals – anomalies in red")
ax.set_ylabel("Residual (%)")
ax.legend(); plt.tight_layout()

# ------------------------------------------------------------
# 2 C – build a CSV of the flagged hours
# ------------------------------------------------------------
csv_cols = ["Timestamp", "Residual_%", "OutdoorTemp_–4h", "HVAC_Mode"]

# hourly HVAC-mode string (first non-NA per hour)
hvac_raw = (merged_df[merged_df["navName"] == "HVAC Mode"]["value"]
            .groupby(pd.Grouper(freq="h"))
            .first())

out_df = pd.DataFrame({
    "Timestamp"      : anomalies.index,
    "Residual_%"     : anomalies.values.round(1),
    "OutdoorTemp_–4h": (wide["Outside Temperature"]
                           .shift(4)                       # –4 h because OA leads demand
                           .reindex(anomalies.index)
                           .round(1)),
    "HVAC_Mode"      : hvac_raw.reindex(anomalies.index)
})

out_df.to_csv("cooling_anomalies_percent.csv", index=False)
print("✓ Exported", len(out_df), "rows  →  cooling_anomalies_percent.csv")
```

* **Bias toward positive residuals**  
  Most flagged points sit **above zero**, meaning the controller frequently calls for **more cooling % than the weather-only model predicts**. Typical root-causes: leaking cooling valve, stages mis-sequencing, or aggressive set-point overrides.


* **A few extreme negatives**  
  The scattered deep dips may be short data gaps or hours when cooling demand was clamped off. Worth cross-checking against alarm / override logs.


* **Anomaly rate ≈ 3 % (224 / 7 411 hrs)**  
  Acceptable for exploration. If we need fewer alerts, raise the threshold (e.g. |z| > 6) or add occupancy & set-points to tighten the baseline.

#### What the triage tells us

* **“Off” but still cooling**

  Classic sign of cooling bleed-through: leaking valve, compressor bypass, or an override that didn’t flip the mode flag.

* **“Cool” mode looks mostly fine**  
  24 high and 15 low residuals; a balanced spread suggests these are weather spikes or airflow hiccups, not a systematic fault.

* **“Heat” mode barely shows up**  
  Just one low-residual hour, as expected—active heating suppresses the cooling-demand percentage.

* **Severity**  
  All anomalies fall into the 10-20 % (35 hrs) or 20-50 % (189 hrs) bands—none in the mild (≤ 5 %) or extreme (≥ 50 %) buckets. So we’re catching medium-to-large deviations without drowning in noise.
